```python
def foo_some(lis, num):
    low = 0
    hight = len(lis) - 1
    while low <= hight:
        mid = (low + hight) // 2
        guess = lis[mid]
        if guess > num or guess == num and lis[mid-1] == num:
            hight = mid - 1
        elif guess == num:
            return lis[mid-1], mid - 1
        elif guess < num < lis[mid + 1]:
            return lis[mid], mid
        elif guess < num:
            low = mid + 1
```

```python
%%time
print(foo_some(list_, n))
```

```python
import json

purchases_dict = {}

with open('purchase_log.txt', encoding='utf-8') as purchases:
    for line in purchases:
        line = json.loads(line)
        key_ = line['user_id']
        purchases_dict.setdefault(key_, [])
        purchases_dict[key_].append(line['category'])
```

```python
%%time
with open('visit_log.csv', encoding='utf-8') as f_log, open('funnel.csv', 'w', encoding='cp1251') as f_funnel:
    for line in f_log:
        b = line.strip().split(',') #список c ID[0] и source[1]
        a = purchases_dict.get(b[0]) # В переменную а записали значение категории из словаря, если ключа в словаре нет в а будет None
        if a != None:
            a = ';'.join(a)
            b.append(a)
            b = ','.join(b)
            f_funnel.write(b)
```

```python
def find_uniq(arr):
    
    return list(map(lambda el: el if arr.count(el) == 1 else , arr))[0]
```
