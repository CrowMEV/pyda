**Задание 1**

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
import pandas as pd
import numpy as np
```

```python
ratings = pd.read_csv('ratings.csv')
```

```python
ratings.head()
```

```python
ratings = ratings.groupby('userId').agg(['count', 'min', 'max'])
```

```python
def classify(row):
    return row['timestamp']['max'] - row['timestamp']['min']   
```

```python
ratings
```

```python
ratings['lifetime'] = ratings.apply(classify, axis = 1)
ratings
```

```python
ratings[ratings['rating']['count'] > 100]['lifetime'].astype('timedelta64[s]').mean()
```

```python
(ratings
 .groupby('userId')
 .agg(
     user_live=('timestamp', np.ptp),
     rating_count=('rating', 'count'),
 )
 .query('rating_count > 100')
 ['user_live']
 .astype('timedelta64[s]')
 .mean()
)
```

```python
ratings = ratings.groupby('userId').agg({'rating': 'count',
                                 'timestamp': lambda x: x.max() - x.min()})
```

```python
ratings[ratings.rating > 100]['timestamp'].astype('timedelta64[s]').mean()
```

**Задание 2**

Дана статистика услуг перевозок клиентов компании по типам (см. файл с кодом занятия). Необходимо сформировать две таблицы:
таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
аналогичную таблицу по типам выручки с указанием адреса клиента
Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
len(rzd)
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
len(auto)
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
len(air)
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
len(client_base)
```

```python
rides = rzd.merge(auto, on = 'client_id', how = 'outer').merge(air, on = 'client_id', how = 'outer')
rides
```

```python
rides.fillna(0, inplace = True) 
rides = rides.astype(int) 
```

```python
rides
```

```python
rides_clients = client_base.merge(rides, on = 'client_id', how = 'inner')
rides_clients
```
