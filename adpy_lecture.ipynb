```python
def get_dataframe(year):
    df_ = pd.read_csv(f'names/yob{year}.txt', names=['Names', 'Gender', 'Count'])
    df_['Year'] = year
    return df_
```

```python
def concat_df(*years):
    df_list = []
    for year in years:
        df_list.append(get_dataframe(year))
    return pd.concat(df_list)
```

```python
def get_top_names(*years):
    return concat_df(*years).groupby(['Names', 'Gender']).max().sort_values(by='Count', ascending=False).iloc[:3].reset_index()['Names'].to_list()
```

```python
get_top_names(1900, 1950, 2000)
```

```python
get_top_names(1880)
```

```python
def count_dynamics(*years):
    df_ = concat_df(*years).groupby(['Gender', 'Year']).sum().unstack()
    return {'F': df_.loc['F'].to_list(), 'M': df_.loc['M'].to_list()}
```

```python
count_dynamics(1900, 1950, 2000)
```
