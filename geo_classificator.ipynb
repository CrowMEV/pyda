```python
import pandas as pd
```

```python
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def some_func(keyword):
    for region, cities_list in geo_data.items():
        for city in cities_list:
            if city in keyword:
                return region
    return 'undefined'

```

```python
geo_df = pd.read_csv('keywords.csv')
```

```python
geo_df['region'] = geo_df['keyword'].apply(some_func)
geo_df
```

```python
x = geo_df[geo_df.region == 'Центр']
x
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
movies_data = ratings.merge(movies, on='movieId', how='left')
movies_data
```

```python
movies_data = movies_data[movies_data.rating == 5.0]
movies_data
```

```python
movies_data = movies_data.groupby('title')[['rating']].count().sort_values('rating', ascending = False)
```
